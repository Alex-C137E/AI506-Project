```python
import pandas as pd
import numpy as np
import re
import networkx as nx
from csv import reader
from sklearn.linear_model import LogisticRegression
```

```python
# Import the embedding
df = pd.read_csv('Embedding/Embp1q2.csv', delimiter=",")
```

```python
# Create an embedding dictionary from the file
embedding_dict = {}
for index, row in df.iterrows():
    ingredient_id = row[1]
    embedding_string = row[2]
    embedding_formatted = re.findall(r'[\d|\.]+', embedding_string)
    embedding = np.asarray(embedding_formatted, dtype=float)
    embedding_dict[ingredient_id] = embedding
    
embedding_dict
```
